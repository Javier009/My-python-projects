```python
def loneliest_number(initial_number,end_number):
    #initial_number = 0
    #end_number = 100
    list_to_analyse = list(range(initial_number, end_number+1))
    result_tuples = []

    for n in list_to_analyse:
        temp_results = []
        if n == 0 or n == 1:
            pass
        else:
            left_block = list(range(1,n))
            right_block = [n]

            ## LEFT BLOCK
            temp_left_block = []
            for i in left_block:
                left_pre_results = []
                left_prime_check_list = list(range(2,i))
                for d in left_prime_check_list:
                    if i%d == 0:
                        left_pre_results.append(1)
                    else:
                        left_pre_results.append(0)
                if sum(left_pre_results) == 0:
                    temp_left_block.append((n,i,n-i))
                else:
                    pass

            left_min_value = min([i[2] for i in temp_left_block])
            for i in [i for i in temp_left_block if i[2] == left_min_value]:
                temp_results.append(i)


            ## RIGHT BLOCK
            is_prime = False
            aditional = right_block[-1]+1

            while is_prime == False:

                right_prime_check_list = list(range(2,aditional))
                right_pre_results = []

                for r in right_prime_check_list:
                    if aditional%r == 0:
                        right_pre_results.append(1)
                    else:
                        right_pre_results.append(0)

                if sum(right_pre_results) == 0:
                    temp_results.append((n,aditional,aditional-n))
                    is_prime = True
                    break

                else:
                    right_block.append(aditional)
                    aditional += 1

            min_value = sum(min([i[2]] for i in temp_results))
            tuples_to_consider = [i for i in temp_results if i[2] == min_value]

            for tup in tuples_to_consider:
                result_tuples.append(tup)

    # FINAL SELECTION OF NUMBERS 

    max_final = sum(max([i[2]] for i in result_tuples))
    final_list_of_tuples = [i for i in result_tuples if i[2] == max_final]

    results = []
    for i in final_list_of_tuples:
        text_to_add = str(final_list_of_tuples.index(i))
        dictionary = {f'number_{text_to_add}' : i[0],
                      f'distance_{text_to_add}' : i[2], 
                      f'closest_{text_to_add}' : i[1]
                     }
        results.append(dictionary)

    for i in results:
        print(i)
```
