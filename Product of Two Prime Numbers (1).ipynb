### Product of Two Prime Numbers, level: expert

https://edabit.com/challenge/437h8sNsWAPCcMRSg

Write a function that returns True if the given number num is a product of any two prime numbers.

Examples
product_of_primes(2059) ➞ True

29*71=2059

product_of_primes(10) ➞ True

2*5=10

product_of_primes(25) ➞ True

5*5=25

product_of_primes(999) ➞ False

There are no prime numbers.

Notes
num is always greater than 0.
0 and 1 aren't prime numbers.

```python
def prime_product(num):
    
    results_tuples = []

    for i in range(1, num+1):

        divisor = i
        quotient = num/i
        residual = num%i

        if residual == 0:

            quotient = int(quotient)
            divisor = int(divisor)
            quotient_list = []
            divisor_list = []

            for q in range(2,int(quotient)):

                if int(quotient) % int(q) != 0:
                    quotient_list.append('Is Prime')
                else:
                    quotient_list.append('Not Prime')

            for d in range(2,int(divisor)):

                if int(divisor) % int(d) != 0:
                    divisor_list.append('Is Prime')
                else:
                    divisor_list.append('Not Prime')

        else:
            pass


        if "Not Prime" not in quotient_list and "Not Prime" not in divisor_list:
            results_tuples.append((divisor,quotient))

        else:
            pass

    final_list = [ i for i in results_tuples if '.' not in str(i[1]) and (str(i[1]) != '1' and str(i[0]) != '1' ) ]
    
    if len(final_list) > 0:
        return (True, final_list[0])
    else:
        return False
```

```python
for i in range(1,5000):
    print(i, "--->" ,prime_product(i))
```
