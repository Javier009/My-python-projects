#Write a function that accepts a positive integer between 0 and 999 inclusive and returns a string representation of that integer written in English.

#Examples
#num_to_eng(0) ➞ "zero"

#num_to_eng(18) ➞ "eighteen"

#num_to_eng(126) ➞ "one hundred twenty six"

#num_to_eng(909) ➞ "nine hundred nine"
#Notes
#There are no hyphens used (e.g. "thirty five" not "thirty-five").
#The word "and" is not used (e.g. "one hundred one" not "one hundred and one").

```python
units = {'0' : ['zero',''],
         '1' : 'one',
         '2' : 'two',
         '3' : 'three',
         '4' : 'four',
         '5' : 'five',
         '6' : 'six',
         '7' : 'seven',
         '8' : 'eight',
         '9' : 'nine'
        }

decens = {'0' : '',
          '1' : {'10': 'ten', '11': 'eleven', '12': 'twelve', '13': 'therteen', '14': 'fourteen', 
                 '15': 'fifteen', '16': 'sixteen', '17': 'seventeen', '18': 'eighteen','19': 'nineteen'},
          '2' : 'twenty',
          '3' : 'thirthy',
          '4' : 'fourty',
          '5' : 'fifty',
          '6' : 'sixty',
          '7' : 'seventy',
          '8' : 'eighty',
          '9' : 'ninety'
         }

centens = {'0' : '',
           '1' : 'one hundred',
           '2' : 'two hundred',
           '3' : 'three hundred',
           '4' : 'four hundred',
           '5' : 'five hundred',
           '6' : 'six hundred',
           '7' : 'seven hundred',
           '8' : 'eight hundred',
           '9' : 'nine hundred'
         }
```

```python
def number_to_string(number_digit):
    number = str(number_digit)

    if len(number) >3:
        return 'Please type a number between 0-999'

    elif len(number) == 1:

        if number == '0':
            result = units[number][0]
            return result
        else:
            result = units[number]
            return result

    elif len(number) == 2:

        if number[0] == '0' and number[1] == '0':
            result = units[number][0]
            return result
        elif number[0] == '1':
            result = decens['1'][number]
            return result
        else:
            if number[1] == '0':
                result = decens[number[0]]
                return result
            else:
                result = decens[number[0]] + ' ' + units[number[1]]
                return result

    else:

        if number[1] == '0' and number[2] == '0':
            result = centens[number[0]] + ' ' + decens[number[1]] + ' ' + units[number[2]][1]
            return result
        
        elif number[1] == '0':
            result = centens[number[0]] + ' ' +  units[number[2]]
            return result

        elif number[1] == '1':
            decens_to_concat = number[1] + number[2] 
            result = centens[number[0]] + ' ' + decens['1'][decens_to_concat]
            return result

        else:
            
            if number[2] == '0':
                result = centens[number[0]] + ' ' + decens[number[1]] + ' ' + units[number[2]][1]
                return result
            else:
                result = centens[number[0]] + ' ' + decens[number[1]] + ' ' + units[number[2]]
                return result
                
```

```python
print(number_to_string(659))
```

```python
for i in range(0,1000):
    print(number_to_string(i))
    
```
