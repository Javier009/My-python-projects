Area of Overlapping Rectangles
Create a function that returns the area of the overlap between two rectangles. The function will receive two rectangles, each with the coordinates of the lower left corner followed by the width and the height rect = [x, y, width, height].

level: expert

link: https://edabit.com/challenge/NrWECd98HTub87cHq


```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
def intersection_area(rec_1, rec_2):
    #rec_1 = [3,6,10,8]
    #rec_2 = [6,8,5,7]

    r1x1 = rec_1[0]
    r1x2 = rec_1[0] + rec_1[2]

    r2x1 = rec_2[0]
    r2x2 = rec_2[0] + rec_2[2]

    r1y1 = rec_1[1]
    r1y2 = rec_1[1] + rec_1[3]

    r2y1 = rec_2[1]
    r2y2 = rec_2[1] + rec_2[3]

    ### Review the intesections for x & y by creating sets with the coordenates:

    #Getting the X sets (coordeates) for each recatangle to see if there is X intersection
    x_set_rec_1 = set([i for i in range(r1x1, r1x2+1)])
    x_set_rec_2 = set([i for i in range(r2x1, r2x2+1)])

    X_intersection = x_set_rec_1.intersection(x_set_rec_2)

    #Getting the X sets (coordeates) for each recatangle to see if there is X intersection
    y_set_rec_1 = set([i for i in range(r1y1, r1y2+1)])
    y_set_rec_2 = set([i for i in range(r2y1, r2y2+1)])

    Y_intersection = y_set_rec_1.intersection(y_set_rec_2)

    #Calculating the area of 3 recatnagles: rec_1, rec_2 and the intersection

    if len(X_intersection) > 0 and len(Y_intersection) > 0:
        area_rec_1 = rec_1[2] * rec_1[3]
        area_rec_2 = rec_2[2] * rec_2[3]
        area_intersection = (max(X_intersection) - min(X_intersection)) * (max(Y_intersection) - min(Y_intersection)) 
    else:
        area_rec_1 = rec_1[2] * rec_1[3]
        area_rec_2 = rec_2[2] * rec_2[3]
        area_intersection = None


    if area_intersection is not None:
        fig = plt.figure(figsize=(20,20))
        ax = fig.add_subplot(111)

        rect1 = matplotlib.patches.Rectangle((r1x1, r1y1),
                                             rec_1[2], rec_1[3],
                                             color ='gray')

        rect2 = matplotlib.patches.Rectangle((r2x1, r2y1),
                                             rec_2[2], rec_2[3],
                                             color ='gray')

        rect3 = matplotlib.patches.Rectangle((min(X_intersection), min(Y_intersection)),
                                             max(X_intersection) - min(X_intersection), max(Y_intersection) - min(Y_intersection),
                                             color ='red',
                                             label = area_intersection)

        ax.add_patch(rect1)
        ax.add_patch(rect2)
        ax.add_patch(rect3)

        plt.xlim([min(r1x1,r1x2, r2x1,r2x2)-2, max(r1x1,r1x2, r2x1,r2x2)+2])
        plt.ylim([min(r1y1,r1y2, r2y1,r2y2)-2, max(r1y1,r1y2, r2y1,r2y2)+2])
        plt.title('Area of intersection between recatangles', fontsize = 30)
        plt.text(min(X_intersection)+0.1 , (min(Y_intersection)+max(Y_intersection))/2, f'Area of intersection: {area_intersection} cm^2', weight="bold", fontsize=10)  
        plt.show()

    else:
        fig = plt.figure(figsize=(20,20))
        ax = fig.add_subplot(111)

        rect1 = matplotlib.patches.Rectangle((r1x1, r1y1),
                                             rec_1[2], rec_1[3],
                                             color ='gray')

        rect2 = matplotlib.patches.Rectangle((r2x1, r2y1),
                                             rec_2[2], rec_2[3],
                                             color ='gray')
        ax.add_patch(rect1)
        ax.add_patch(rect2)
        plt.title('Area of intersection between recatangles', fontsize = 30)
        plt.xlim([min(r1x1,r1x2, r2x1,r2x2)-2, max(r1x1,r1x2, r2x1,r2x2)+2])
        plt.ylim([min(r1y1,r1y2, r2y1,r2y2)-2, max(r1y1,r1y2, r2y1,r2y2)+2])
        plt.show()
```
