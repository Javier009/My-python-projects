```python
#The Josephus Problem
#This classic problem dates back to Roman times. There are 41 soldiers arranged in a circle. Every third soldier is to be killed by their captors, continuing around the circle until only one soldier remains. He is to be freed. Assuming you would like to stay alive, at what position in the circle would you stand?

#Generalize this problem by creating a function that accepts the number of soldiers n and the interval at which they are killed i, and returns the position of the fortunate survivor.

#Examples
#josephus(41, 3) ➞ 31

#josephus(35, 11) ➞ 18

#josephus(11, 1) ➞ 11

#josephus(2, 2) ➞ 1
# Level: expert
#link : https://edabit.com/challenge/Mb8KmicGqpP3zDcQ5
```

```python
def josephus(s,k):
    soliders = s
    interval_of_kills = k
    interval_of_kills_list = [i for i in range(1,interval_of_kills+1) ]
    soldiers_list = [i for i in range(1,soliders+1) ]
    soldiers_list
    soldiers_list_replication = soldiers_list
    round_iterator = 1 

    while len(soldiers_list_replication) > 1:

        groups = [interval_of_kills_list[i:i+len(soldiers_list_replication)] for i in range(0, len(interval_of_kills_list), len(soldiers_list_replication))]
        
        for list_groups in groups:
            if len(list_groups) < len(soldiers_list_replication):
                to_add = len(soldiers_list_replication)-len(list_groups)
                for i in range (1, to_add+1):
                    list_groups.append(0)
            else:
                pass

        for l in groups:
            assigment = [(s , l[soldiers_list_replication.index(s)]) for s in soldiers_list_replication]

            for s in assigment:
                if s[1] == interval_of_kills:
                    soldier_to_kill = s[0]
                    rotation = [r for r in soldiers_list_replication if soldiers_list_replication.index(r) < soldiers_list_replication.index(soldier_to_kill)]
                    print(f'Solider to kill round {round_iterator}: {soldier_to_kill}')
                    round_iterator +=1
                    for r in rotation:
                        soldiers_list_replication.pop(soldiers_list_replication.index(r))
                        soldiers_list_replication.append(r)

                    soldiers_list_replication.pop(soldiers_list_replication.index(soldier_to_kill))

                else:
                    pass
    else:
        print(f'To survive stay in posistion number: {soldiers_list_replication[0]}')
    
```
